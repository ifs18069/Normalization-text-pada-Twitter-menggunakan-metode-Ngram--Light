```python
#menerapkan word_tokenize untuk semua catatan, dan membuat kolom baru di DataFrame dataset kami. 
#Setiap entri akan menjadi daftar kata-kata. Di sini juga akan menghapus kata / karakter non alfanumerik
#(seperti angka dan tanda baca)
def identify_tokens(row):
    text= row['text']
    tokens = nltk.word_tokenize(text)
    # hanya mengambil kata(tidak punctuation)
    token_words = [w for w in tokens if w.isalpha()]
    return token_words

data['words'] = data.apply(identify_tokens, axis=1)
```
